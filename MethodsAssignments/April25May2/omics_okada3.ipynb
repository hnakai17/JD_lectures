```r
#Code
library(e1071)
moment <- function(x){
  mu <- mean(x)
  sigma2 <- var(x)
  sk <- skewness(x)
  kr <- mean(((x-mu)/sd(x))^4) - 3
  result <- c(mu,sigma2,sk,kr)
  names(result) <- c("mean","var","skewness","kutrosis")
  return(result)
}

#Generate random values
set.seed(1000)
x <- rnorm(10000,10,4)

#skeness and kurtosis is essentially zero
#essentially zero means that Although it differs slightly from zero depending on how to select specimens, it varies around zero
moment(x)
hist(x)

#another random value set
x2 <-  rnorm(10000,-2,2)
x3 <- c(x,x2)
hist(x3)

#skeness and kurtosis is not zero
moment(x3)
```
