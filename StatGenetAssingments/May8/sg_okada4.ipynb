```r
#Generate positive-definite matrix by generating data matrix and making its covariance matirix
n <- 10
mat <- matrix(rnorm(n^2),n,n)
cov_mat <- var(mat)
barplot(eigen(cov_mat)$values)

#by generating a diagonal matrix whose diagonal elements >0 and sandwich it with rotation matrix and its inverse.
#rotaion matrix satisfy Orthogonal matrix
#Input is eigen values
diag_mat <- diag(c(1,2))
sita <- pi/3
rot_mat <- matrix(c(cos(sita),sin(sita),-sin(sita),cos(sita)),2,2)
pdm <- rot_mat %*% diag_mat %*% solve(rot_mat)
barplot(eigen(pdm)$values)
```
